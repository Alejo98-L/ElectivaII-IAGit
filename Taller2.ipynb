```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data_importada = pd.read_csv('./mi_df_minable.csv',sep=',',decimal='.')
df_data_csv = pd.DataFrame(data_importada)


# Graficar con plt
plt.figure(figsize=(12,5))

# Histograma de Valor credito
plt.subplot(1,2,1)
sns.histplot(df_data_csv['Valor_credito'], bins=5, kde=True, color='blue')
plt.title('Valor credito de las personas')
plt.xlabel('Credito')
plt.ylabel('Cantidad')
```

```python

# Graficar con plt
plt.figure(figsize=(12,5))

# Barras de estado civil
plt.subplot(1,2,1)
sns.countplot(x='Estado_civil', data=df_data_csv, palette='viridis')
plt.title('Estado_civil')
plt.xlabel('Estado')
plt.ylabel('Frecuencia')
```

```python
import statistics
# import matplotlib.pyplot as plt

media = statistics.mean(df_data_csv['Valor_credito'])
mediana = statistics.median(df_data_csv['Valor_credito'])
moda = statistics.mode(df_data_csv['Valor_credito'])

print(f"Media: {media}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda}")
```
