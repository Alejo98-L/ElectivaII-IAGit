```python
import pandas as pd

data_importada = pd.read_csv('./2024_hourly_data.csv',sep=',',decimal='.')
dataFrame = pd.DataFrame(data_importada)
```

```python
valoresNulos = dataFrame.isnull().sum()
print(valoresNulos)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.boxplot(x=dataFrame["NH3"])
plt.show()
```

```python
sns.boxplot(x=dataFrame["PM10"])
plt.show()
```

```python
sns.boxplot(x=dataFrame["SO2"])
plt.show()
```

```python
dfCO = dataFrame['CO']
dfNH3 = dataFrame['NH3']
dfNO2 = dataFrame['NO2']
dfOZONE = dataFrame['OZONE']
dfPM10 = dataFrame['PM10']
dfPM2_5 = dataFrame['PM2.5']
dfSO2 = dataFrame['SO2']

dfCO = dfCO.dropna()
dfNH3 = dfNH3.dropna()
dfNO2 = dfNO2.dropna()
dfOZONE = dfOZONE.dropna()
dfPM10 = dfPM10.dropna()
dfPM2_5 = dfPM2_5.dropna()
dfSO2 = dfSO2.dropna()
```

```python
dfCO.to_csv("Calidad_renovada.csv", index=False)
dfNH3.to_csv("Calidad_renovada2.csv", index=False)
dfNO2.to_csv("Calidad_renovada3.csv", index=False)
dfOZONE.to_csv("Calidad_renovada4.csv", index=False)
dfPM10.to_csv("Calidad_renovada5.csv", index=False)
dfPM2_5.to_csv("Calidad_renovada6.csv", index=False)
dfSO2.to_csv("Calidad_renovada7.csv", index=False)
```
